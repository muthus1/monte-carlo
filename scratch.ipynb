```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
data = np.random.standard_normal((5, 100))
```

```python
plt.figure(figsize=(10, 6));
plt.plot(data.T.cumsum(axis=0), label=['line1','line2','line3','line4','line5'])
plt.legend(loc=0);
plt.xlabel('data point')
plt.ylabel('value');
plt.title('random series');
```

```python
plt.figure(figsize=(10, 6))
plt.bar(np.arange(1, 12),
data[0, :11], width=0.5)
```

```python
x = np.arange(len(data.cumsum()))
y = 0.2 * (data.cumsum() ** 2)
rg1 = np.polyfit(x,y,1)
rg3 = np.polyfit(x,y,3)
rg9 = np.polyfit(x,y,9)
plt.figure(figsize = (10,6))
plt.plot(x,y,'r', label = 'data')
plt.plot(x, np.polyval(rg1, x), 'b--', label='linear')
plt.plot(x, np.polyval(rg3, x), 'b-.', label='cubic')
plt.plot(x, np.polyval(rg9, x), 'b:', label='9th degree')
plt.legend(loc=0);
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Fit a linear polynomial (degree 1) using np.polyfit
coefficients = np.polyfit(x, y, 1)
print(coefficients)
```

```python
predicted_y = np.polyval(coefficients, x)
print(predicted_y)
```
